```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import sklearn
```

# Load in the Training Data

```python
train = np.loadtxt('sample_data/mnist_train_small.csv', delimiter=',')
train_data = train[:,1:]
train_labels = train[:,0].astype(np.int16)
```

# Load in the Testing Data

```python
# Load test data
test = np.loadtxt('sample_data/mnist_test.csv', delimiter=',')
test_data = test[:,1:]
test_labels = test[:,0].astype(np.int16)
```

## Your Task!

I implemented a quick linear classifier. You need to beat my model (91%)

```python
from sklearn import linear_model, metrics

clf = linear_model.LogisticRegression()

clf.fit(train_data, train_labels)
```

# See How Well You Did

```python
predicted = clf.predict(test[:,1:])

print(
    f"Classification report for classifier {clf}:\n"
    f"{metrics.classification_report(test[:,0], predicted)}\n"
)

disp = metrics.ConfusionMatrixDisplay.from_predictions(test[:,0], predicted)
disp.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{disp.confusion_matrix}")

plt.show()
```
